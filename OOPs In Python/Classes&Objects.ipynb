# OOPS - Object oriented Programming


```python
# Using OOPs creating student's records

#class: blueprint or template
class Student:
    name="divya"
    sem=8

# object: instance of a class
student1=Student()
print(student1.name,student1.sem)

student2=Student() #here we can't give any arguments"
print(student2.name,student2.sem) 
```

## __init__ :- initialize ( helps to intialize the values)  ( always takes self as one of the parameter)

## __self__:- the self parameter is a reference to the current instance of the class and is used to access the variables that belong to the class   
( self is used to make the connection with the class to the object)




```python
#class: blueprint or template
class Student:
    def __init__(self,full_name,class_grade):   # different name
        self.name=full_name
        self.grade=class_grade

# object: instance of a class
student1=Student("Madhav",10)
print(f'{student1.name} is of {student1.grade} grade')

student2=Student("Vishakha",12)
print(f'{student2.name} is of {student2.grade} grade')
```

```python
#class: blueprint or template
# __init__ method:- constructor, value initialize-fix
# self:- reference , connection build between class and object -fix

class Student:                        # Student class
    def __init__(self,name,grade):    # Method /function
        self.name=name                # Attribute / properties
        self.grade=grade              # Attribute / properties

# object: instance of a class
student1=Student("Madhav",10)
print(f'{student1.name} is of {student1.grade} grade')

student2=Student("Vishakha",12)
print(f'{student2.name} is of {student2.grade} grade')
```

```python
#class: blueprint or template
# __init__ method:- constructor, value initialize-fix
# self:- reference , connection build between class and object -fix

class Student:  # Student class
    def __init__(self,name,grade,percentage):    # Method /function
        self.name=name       # Attribute / properties
        self.grade=grade     # Attribute / properties
        self.percentage=percentage   # Attribute

    def student_details(self):    # method
        print(f'{self.name} is of {self.grade} grade with {self.percentage}')

# object: instance of a class
student1=Student("Madhav",10,89.0)
# print(f'{student1.name} is of {student1.grade} grade')

student2=Student("Vishakha",12,99.0)
# print(f'{student2.name} is of {student2.grade} grade')

print(student1.name)
student1.student_details()
student2.student_details()

print(student1.__dict__)    # gives the details of student1 in form of dictionary
```

## OOP Operations in class and Objects

```python
# modify Object Property 

print(student1.percentage)
student1.percentage=78.5
print(student1.percentage)
```

```python
# Delete Object property

print(student1.__dict__)
del student1.percentage
print(student1.__dict__)
```

```python
# delete the object

print(student1)
del student1
# print(student1)   # will give error:-  name 'student1' is not defined
```

```python
class Student:  # Student class
    def __init__(self,name,grade,percentage,team):    # Method /function
        self.name=name       # Attribute / properties
        self.grade=grade     # Attribute / properties
        self.percentage=percentage   # Attribute
        self.team=team

    def student_details(self):    # method
        print(f'{self.name} is of {self.grade} grade with {self.percentage} is in {self.team}')

team1="A"
team2="B"


# object: instance of a class
student1=Student("Madhav",10,89.0,team1)
# print(f'{student1.name} is of {student1.grade} grade')

student2=Student("Vishakha",12,99.0,team2)
# print(f'{student2.name} is of {student2.grade} grade')

print(student1.team)
student1.student_details()
student2.student_details()
```
