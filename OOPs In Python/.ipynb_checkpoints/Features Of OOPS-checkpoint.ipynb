# ABSTRACTION
### ABSTRACTION:- Hiding Unnecessary details from class , methods 
### Abstraction in Python: Hiding unnecessary details
### Abstraction hides implementation details and shows only the relevant functionality to the user.

```python
# Abstraction: - Hiding Unnecessary details from class , methods 

class Student:  # Student class
    def __init__(self,name,grade,percentage):    # Method /function
        self.name=name       # Attribute / properties
        self.grade=grade     
        self.percentage=percentage   

# here we are using abstraction 
# where we are giving percentage a raise by 2 and we don't want the users to know so this will hide the information
    def student_details(self):    # method
        print(f'{self.name} is of {self.grade} grade with {self.percentage +2}')  # hidden from users

# object: instance of a class
student1=Student("Madhav",10,89.0)
student2=Student("Vishakha",12,90.0)

student1.student_details()
student2.student_details()
```

# ENCAPSULATION
### Encapsulation:-  Restrict access to certain attributes or methods to protect data and enforce controlled access
### Encapsulation in Python: Restricting direct access to attributes & methods
### Encapsulation restricts access to certain attributes or methods to protect the data and
### enforce controlled access.




```python
# Encapsulation :-  Restrict access to certain attributes or methods to protect data and enforce controlled access
# we can see the private attribute or methods  by making a new function or method 


class Student:  # Student class
    def __init__(self,name,grade,percentage):    # Method /function
        self.name=name       # Attribute / properties
        self.grade=grade     
        self.__percentage=percentage    # Double underscore limits access

    def get_percentage(self):
        return self.__percentage
    
    def student_details(self):    # method
        print(f'{self.name} is of {self.grade} grade with {self.percentage +2}')

# object: instance of a class
student1=Student("Madhav",10,89.0)
student2=Student("Vishakha",12,90.0)

# print(student1.percentage)  # gives error :- 'Student' object has no attribute 'percentage'

print(student1.get_percentage())

```

# INHERITANCE:-
### Inheritance in Python: Reusing Parentâ€™s prop & methods
### Inheritance (parent-child), allows one class (child) to reuse the properties and methods of
### another class (parent). This avoids duplication and helps in code reuse.
### Inheritance:- Allows one class(child) to reuse the properties and methods of another class(parent).

```python
# INHERITANCE:- Allows one class(child) to reuse the properties and methods of another class(parent).  


# Property inheritance:---
# parent class
class Student:  # Student class
    def __init__(self,name,grade,percentage):    # Method /function
        self.name=name       # Attribute / properties
        self.grade=grade     
        self.percentage=percentage   
    
    def student_details(self):    # method
        print(f'{self.name} is of {self.grade} grade with {self.percentage +2}')

# object: instance of a class
student1=Student("Madhav",10,89.0)
student2=Student("Vishakha",12,90.0)



#Child Class
class GraduateStudent(Student):   # GraduateStudent class inherit prop and methods from Student parent class
    def __init__(self,name,grade,percentage,stream):  # old parameters from parent class and new parameters from new child class
        super().__init__(name, grade, percentage)  # call parent class initializer(init)
        self.stream=stream

# Object 
grad_student1=GraduateStudent("Divya",12,89.0,"PCM")
grad_student2=GraduateStudent("divya",10,97.0,"SSC")
print(grad_student1.stream)
print(grad_student1.percentage)
print(grad_student1.__dict__)
print(grad_student2.__dict__)
```

```python
# INHERITANCE:- Allows one class(child) to reuse the properties and methods of another class(parent).  


# Method Inheritance
# parent class
class Student:  # Student class
    def __init__(self,name,grade,percentage):    # Method /function
        self.name=name       # Attribute / properties
        self.grade=grade     
        self.percentage=percentage   
    
    def student_details(self):    # method
        print(f'{self.name} is of {self.grade} grade with {self.percentage +2}')

# object: instance of a class
student1=Student("Madhav",10,89.0)
student2=Student("Vishakha",12,90.0)



#Child Class
class GraduateStudent(Student):   # GraduateStudent class inherit prop and methods from Student parent class
    def __init__(self,name,grade,percentage,stream):  # old parameters from parent class and new parameters from new child class
        super().__init__(name, grade, percentage)  # call parent class initializer(init)
        self.stream=stream

    def student_details(self):
        super().student_details()   # Method inherit from parent class
        print(f'stream is {self.stream}')

# Object 
grad_student1=GraduateStudent("Divya",12,89.0,"PCM")
grad_student2=GraduateStudent("divya",10,97.0,"SSC")
# print(grad_student1.stream)
# print(grad_student1.percentage)
# print(grad_student1.__dict__)
# print(grad_student2.__dict__)


grad_student1.student_details()
```

# POLYMORPHISM 
### Polymorphism in Python: Same method but different output
### Polymorphism allows methods in different classes to have the same name but behave
### differently depending on the object.
### Polymorphism:-Allows methods in different class to have same name but different behaviour depending on objects

```python
# POLYMORPHISM:- Allows methods in different class to have same name but different behaviour depending on objects



# parent class
class Student:  # Student class
    def __init__(self,name,grade,percentage):    # Method /function
        self.name=name       # Attribute / properties
        self.grade=grade     
        self.percentage=percentage   
    
    def student_details(self):    # method
        print(f'{self.name} is of {self.grade} grade with {self.percentage +2}')

# object: instance of a class
student1=Student("Madhav",10,89.0)
student2=Student("Vishakha",12,90.0)



#Child Class
class GraduateStudent(Student):  
    def __init__(self,name,grade,percentage,stream):  
        super().__init__(name, grade, percentage)  # call parent class initializer(init)
        self.stream=stream

    def student_details(self):  #method
        print(f'{self.name} is in {self.grade} with {self.percentage} % from stream {self.stream}' ,"(Same method with different output)")
      


# Object Student class
student1=Student("Madhav",10,89.0)

# Object  GraduateSTudent Class
grad_student1=GraduateStudent("Divya",12,89.0,"PCM")


student1.student_details()
grad_student1.student_details()
```
