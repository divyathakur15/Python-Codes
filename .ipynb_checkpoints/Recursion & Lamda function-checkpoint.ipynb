# RECURSION is Technique where a function calls itself to solve small problems. It is often used for
# problems that can be broken dowm into smaller (similar subproblems like calculatiing factorial ,fibonacci and 
# travasal data like graph and tree)

# The Recursion function has two component 
# 1) Base case   (terminating case)
# 2) recursion case ( where fucntion calls itself)

```python
# write a Recursion Function that returns sum of first n natural number
def sum(n):
    if n==1:
        return 1
    else:
        return n+sum(n-1)
```

```python
sum(10)
```

```python
# write a python function that return factorial of n number
def fact(n):
    if n==0 or n==1:
        return 1
    else:
        return n*fact(n-1)
```

```python
fact(5)
```

```python
# Write a Python function To return The fibonacci series
# f(n)=f(n-1)+f(n-2)
def fibo(n):
    if n==1:
        return 0
    elif n==2:
        return 1
    else:
        return fibo(n-1)+fibo(n-2)
```

```python
n=int(input("Enter How Many Terms:"))
for i in range(1,n+1):
    print(fibo(i),end=" ")
```

```python
def median(l):
    l1=[]
    a=len(l)
    for i in range(len(l)):
        l1=sorted(l)
        if len(l) %2!=0:
            median=l1[a//2]
        else:
            median=(l1[a//2-1]+l1[a//2])
        return median

            
       
```

```python
median([1,2,3,4,5,1,1,3])
```

```python
def mean_median_mode(l):
    sum=0
    count=0
    for i in l:
        sum=sum+i
        count=count+1
    mean=sum/count
    sorted_list=sorted(l)
    n=len(l)
    if n%2!=0:
        median=sorted_list[n//2]
    else:
        median=(sorted_list[n//2-1] + sorted_list[n//2])/2

    d={}
    for i in l:
        if i in d:
            d[i]+=1
        else:
            d[i]=1

        max_value=max(d.values())
        for i ,j in d.items():
            if j==max_value:
                mode=i
        return mean, median , mode 
```

```python
mean_median_mode([1,2,3,4,5,4,1,2,1,2])
```

# High Order function
# map (function,iterables)
# fucntion like lambda and iterables like list ,tuples , set

```python
l=[1,2,3,4,5]
#output: [1,8,27,64,125]

l1=[]
for i in l:
    l1.append(i**3)
print(l1)

print([i**3 for i in l])

print(list(map(lambda i:i**3,l)))
```

```python
l1=[5,1,2,3,4]
l2=[2,5,4,5,1,]
l=[]

for i, j in zip(l1,l2):
    l.append(i**j)
print(l)
print([i**j for i,j in zip(l1,l2)])

print(list(map(lambda i,j:i**j,l1,l2)))
```
