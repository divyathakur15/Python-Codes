# MERGE SORT:-

Merge Sort is a comparison-based, stable sorting algorithm that follows the divide and conquer technique.
It divides the given array into smaller subarrays, sorts each subarray recursively, and then merges the sorted subarrays to produce the final sorted array.

#### Divide and Conquer Approach
##### Merge Sort works on the principle of Divide, Conquer, and Combine:

1) Divide: <br>
The array is repeatedly divided into two halves until each subarray contains only one element (a single element is always sorted).

2) Conquer:<br>
Each divided subarray is sorted recursively.

3) Combine (Merge):<br>
The sorted subarrays are merged together in a sorted manner to form a larger sorted array.

##### Merge Sort â€“ Normal Steps (Algorithm)

1) Divide the array into two halves.

2) Recursively divide each half until each subarray has only one element.

3) Merge the subarrays by comparing elements and arranging them in sorted order.

4) Continue merging until the full array is sorted.

```python
# ASCENDING ORDER: 
```

```python
# Merging two sorted arrays 
def merge_array(left,right):
    res=[]
    n=len(left)
    m=len(right)
    i,j=0,0
    while i<n and j<m:
        if left[i]<=right[j]:
            res.append(left[i])
            i=i+1
        else:
            res.append(right[j])
            j=j+1

    if i<n:
        while i<n:
            res.append(left[i])
            i=i+1
    if j<m:
        while j<m:
            res.append(right[j])
            j=j+1
    return res
            
    
```

```python
def merge_sort(arr):
    if len(arr)<=1:
        return arr
    mid=len(arr)//2
    left_half=arr[:mid]
    right_half=arr[mid:]
    left_half=merge_sort(left_half)
    right_half=merge_sort(right_half)
    return merge_array(left_half,right_half)
    
```

```python
print(merge_sort([1,3,4,5,2,1,3,4,3,5,4]))
```
