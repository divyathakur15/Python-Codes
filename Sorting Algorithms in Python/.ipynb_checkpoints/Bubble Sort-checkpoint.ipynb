# BUBBLE SORT

### What is Bubble Sort?

#### Bubble Sort is a simple sorting technique where adjacent elements are compared and swapped if they are in the wrong order.
#### After each pass, the largest element “bubbles up” to the end of the array.

 Bubble Sort – Normal Steps (Algorithm)

1) Start from the first element of the array.

2) Compare the current element with the next element.

3) If the current element is greater than the next one, swap them.

4) Move to the next pair and repeat until the end of the array.

5) After one full pass, the largest element is placed at the last position.

6) Repeat the process for the remaining unsorted part of the array.

7) Stop when no swaps are needed or all elements are sorted.

Bubble sort, sometimes referred to as sinking sort, is a simple sorting algorithm that repeatedly steps through the list to be sorted, compares each pair of adjacent items and swaps them if they are in the wrong order. The pass through the list is repeated until no swaps are needed which indicates that the list is sorted.

The algorithm which is a comparison sort is named for the way smaller or larger elements "bubble" to the top of the list. Although the algorithm is simple, it is too slow and impractical for most problems even when compared to insertion sort. Bubble sort can be practical if the input is in mostly sorted order with some out-of-order elements nearly in position.

Bubble sort works on adjacent swaps


### STEPS Of BUBBLE SORT:-
1) starting with the first element (index 0) compare the current element with the next element of the list
2) If the current element is greater than the next element of the list swap them
3) If the current element is less than the next element ,move to the next element.
4) At last we will have two parts, the right part will have the sorted part and left will be unsorted 
5) Repeat step 1

```python
# ACENDING ORDER 
nums=[23,45,33,41,21,2,54,12,10]
n=len(nums)
for i in range(n-2,-1,-1):
    for j in range(0,i+1):
        if nums[j]>nums[j+1]:
            nums[j],nums[j+1]=nums[j+1],nums[j]

print(nums)
```

## TIME COMPLEXITY:-   (AVERAGE/ WORST)
 #### TC=O(N(N+1))/2 
#### ~~ TC= O(N^2)

## SPACE COMPLEXITY:-
#### SC= O(1)

```python
# DESCENDING ORDER 
nums=[23,45,33,41,21,2,54,12,10]
n=len(nums)
for i in range(n-2,-1,-1):
    for j in range(0,i+1):
        if nums[j]<nums[j+1]:
            nums[j],nums[j+1]=nums[j+1],nums[j]

print(nums)
```

 #### BEST CASE TIME COMPLEXITY CODE:

  ##### TC=O(1)

```python
nums=[1,2,3,4,5,6,7]
n=len(nums)
for i in range(n-2,-1,-1):
    is_swap=False
    for j in range(0,i+1):
        if nums[j]>nums[j+1]:
            nums[j],nums[j+1]=nums[j+1],nums[j]
            is_swap=True
    if is_swap==False:
        break;

print(nums)
```
