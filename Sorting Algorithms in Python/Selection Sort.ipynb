# Selection Sort 

### Selection sort is an in placed comparison based algorithm in which the list is divided into two parts, the sorted part at the left end and the unsorted part at the right end.

### Selection Sort is a simple sorting algorithm in which the smallest element is repeatedly selected from the unsorted part of the array and placed at its correct position in the sorted part.

# Steps of Selection Sort :  ( ASCENDING ORDER)
1) Search the list and find out the minimum value , if it is true then change the minimum value and if it is false then do nothing
2) Swap the smallest number to the 0th index
3) we will have two sections the left will have sorted part and the right will have unsorted
4) Repeat the steps for the unsorted part 

```python
# Ascending Order
def selection_sort(nums):
    n=len(nums)
    for i in range(0,n):
        min_index=i
        for j in range(i+1,n):
            if nums[j]<nums[min_index]:
                min_index=j
        nums[i],nums[min_index]=nums[min_index],nums[i]
    return nums
        
print(selection_sort([9,3,5,1,4,2,6]))
```

```python
# Descending Order
def selection_sort(nums):
    n=len(nums)
    for i in range(0,n):
        max_index=i
        for j in range(i+1,n):
            if nums[j]>nums[max_index]:
                max_index=j
        nums[i],nums[max_index]=nums[max_index],nums[i]
    return nums
        
print(selection_sort([9,3,5,1,4,2,6]))

```
