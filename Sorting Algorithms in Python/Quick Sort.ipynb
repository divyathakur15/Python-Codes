# QUICK SORT:- 

### Quick Sort is a comparison-based, divide and conquer sorting algorithm that works by selecting a pivot element, partitioning the array around the pivot, and then recursively sorting the subarrays on both sides of the pivot.

#### Quick Sort â€“ Steps (Algorithm)

1) Choose a pivot element from the array.

2) Partition the array so that:

3) Elements smaller than the pivot are placed on the left.

4) Elements greater than the pivot are placed on the right.

5) Place the pivot in its correct sorted position.

6) Apply Quick Sort recursively to the left subarray.

7) Apply Quick Sort recursively to the right subarray.

8) Repeat the process until all subarrays contain one or zero elements.

9) The array is now sorted.


#### Theory of Quick Sort
#### 1) Divide and Conquer Strategy

Quick Sort follows three main steps:

a) Divide:
Choose a pivot element from the array.

b) Conquer (Partition):
Rearrange elements so that:

Elements smaller than the pivot are placed on the left.

Elements greater than the pivot are placed on the right.

c) Combine:
Recursively apply Quick Sort to the left and right subarrays.


#### 2) Pivot Selection

The pivot can be selected in different ways:

a)First element

b) Last element

c) Middle element

d) Random element

Proper pivot selection improves performance.

#### 3. Partitioning Process

a) Pick a pivot.

b) Compare each element with the pivot.

c) Place elements smaller than pivot to the left and larger to the right.

c) After partitioning, the pivot is placed in its correct sorted position.

#### 4. Recursive Sorting

a) Apply Quick Sort on the left subarray.

b) Apply Quick Sort on the right subarray.

c) Continue until subarrays contain one or zero elements.

```python
class Solution:
    # Function to sort a list using quick sort algorithm.
    def quickSort(self, arr, low, high):
        if low < high:
            # Partition the array and get the pivot index
            p_index = self.partition(arr, low, high)
            # Recursively sort elements before and after partition
            self.quickSort(arr, low, p_index - 1)
            self.quickSort(arr, p_index + 1, high)

    def partition(self, arr, low, high):
        pivot = arr[low]
        i = low
        j = high

        while i < j:
            while i <= high - 1 and arr[i] <= pivot:
                i += 1
            while j >= low + 1 and arr[j] > pivot:
                j -= 1
            if i < j:
                arr[i], arr[j] = arr[j], arr[i]

        arr[low], arr[j] = arr[j], arr[low]
        return j
```

```python
arr = [64, 34, 25, 12, 22, 12, 12, 11, 90]

obj = Solution()
obj.quickSort(arr, 0, len(arr) - 1)

print(f"Sorted array = {arr}")
```
