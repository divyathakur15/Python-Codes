```python
# wap to print the sum of first n even and odd numbers?
n=int(input("ENTER NUMBERS"))
sum_even=0
sum_odd=0
for i in range(1,2*n+1,1):
    if i%2==0:
        sum_even=sum_even+i
    else:
        sum_odd=sum_odd+i
print("EVEN NUMBERS:",sum_even)
print("ODD NUMBERS:",sum_odd)
      
```

```python
# WAP to print number of count of even and odd between the range start =1000 and end=2000

num1=1000
num2=2000
count_even=0
count_odd=0
for i in range(num1 ,num2+1,1):
    if i%2==0:
        count_even+=1
    else:
        count_odd+=1
print("count of even numbers between",num1 ,"and", num2,":",count_even)
print("count of odd numbers between",num1 ,"and", num2,":",count_odd)
```

```python
# wap to print factorial of n number?
n=int(input("enter any number:"))
fact=1
for i in range(1,n+1):
    fact*=i
print("factorial of",n,":",fact)
```

```python
# WAP that simulate a simple ATM .Ask the user for their account balance and  the amount they want to withdraw. 
# If the balance is sufficient ,update the final balance otherwise display an insufficient balance message 

acc_balance=float(input("ENTER ACCOUNT BALANCE:"))
withdrawal_money=float(input("ENTER MONEY TO WITHDRAW:"))
updated_amount=0
if acc_balance>=withdrawal_money:
    updated_amount=acc_balance-withdrawal_money
    print("ACCOUNT BALANCE REMAINING:",updated_amount)
else:
    print("INSUFFICIENT BALANCE IN ACCOUNT")
```
