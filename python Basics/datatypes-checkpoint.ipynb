```python
# STRING - ordered ,sequence of characters 
# single line string is denoted by '' or " " and  multiline string denoted by ''' ''' .
# string is indexed and immutable
a='hello'
print(a)
print(type(a))   # str class 
print(a[0])   # indexed 
a[1]="z"     # immutable    ( 'str' object does not support item assignment )        
print(a)
```

```python
# LIST - collection of elements 
# denoted by[] with comma separated values .
# ordered ,indexed ,mutubale and  heterogenous in nature ,

l=[]   # EMPTY LIST 
print(l)
print(type(l))
```

```python
l=[10,20,"decode",10.5,"bhopal"]
print(l)
print(type(l))

l[1]="DATA"
print(l)

print(l[2][0])
l[2]=[100,200,300]
print(l)
print(l[2][1])
```

```python

# tuple - collection of elements 
# denoted by () with comma separated values
# tuples are ordered ,indexed , heterogeneous and immutable in nature
t=()       # EMPTY TUPLE
print(t)
print(type(t))

```

```python
# Single valued tuple 
t=(10,)
print(t)
print(type(t))
```

```python
t=(10,20,"decode",10.5,"d")
print(t)
print(type(t))

print(t[2])
t[1]="hello"
print(t)        #'tuple' object does not support item assignment
```

```python

#DICTIONARY 
# Collection of elements in a KEY:VALUE pair
#it is denoted by {k:v}
# ordered ,unindexed and mutable in nature 
# keys must be unique and values can be duplicate 

d={}   # empty dictionary
print(d)
print(type(d))
```

```python
d={"name":"xyz","subject": "DA","fees":25000}
print(d)
print(type(d))

d["name"]="DECODE"
print(d)
```

```python
d={"name":"xyz","subject": "DA","fees":25000}
print(d)
d["NAME"]="DECODE"
print(d)
```

```python
d={}
d["NAME"]="DECODE"
print(d)
```

```python
d={"NAME":['abc','def','pqr'],"SUBJECT": ["DA","DS","AI"],"FEES":[24000,25000,34000]}
print(d)

print(d["NAME"][1])

d["NAME"][1]="XYZ"
print(d)
```

```python
# SET - COLLECTION OF ELEMENTS
 # denoted by {} with comma separated values 
# unordered ,unindexed ,heterogeneous and mutable .
 # STORE only UNIQUE VALUES  
s=set()     # EMPTY SET   ---set() con
print(s)
print(type(s))
```

```python
s={ 10,10.5,"HELLO",10,10,10.5,}
print(s)
print(type(s))

s.add(100)      # MUTABLE
print(s)
```

```python
# FROZEN SET--- same as sets      but it is ( immutable) in nature

```

```python
# Create a frozenset from a list
my_list = [1, 2, 3, 2, 4]
my_frozenset = frozenset(my_list)
print(my_frozenset)

# Create a frozenset from a tuple
my_tuple = ("apple", "banana", "cherry")
another_frozenset = frozenset(my_tuple)
print(another_frozenset)
```

```python
fruits = frozenset(["apple", "banana", "orange"])
print(fruits) 
fruits.add("cherry")
print(fruits)
```
