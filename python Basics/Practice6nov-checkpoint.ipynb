```python
# Count Digits of a Number
# Input a number and print how many digits it has (no string conversion).
n=int(input("enter  number"))
count=0
while n!=0:
    count=count+1
    n=n//10
print(count)
```

```python
# Reverse a Number
# Reverse a given number using a while loop.
n=int(input("Enter a num"))
rev=0
while n!=0:
    dig=n%10
    rev=dig+10*rev
    n=n//10
print(rev)
```

```python
# Check Palindrome Number
# Input a number and check if it reads the same backward (ex: 121 → palindrome).
n=int(input("enter a num"))
rev=0
number =n
while n!=0:
    dig=n%10
    rev=dig+rev*10
    n=n//10
if rev==number:
    print("Palindrome")
else:
    print("Not palindrome")
```

```python
# Print All Prime Numbers up to n
# Input n and print all primes from 2 to n.

n=int(input('enter a num'))
for i in range(2,n+1):
    for j in range(2,i):
        if i%j==0:
            break
    else:
        print(i,end=" ")
```
