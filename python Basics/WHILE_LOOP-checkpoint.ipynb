```python
# WAP to print reverse of any number
n=int(input("ENTER ANY NUMBER"))
rev=0
while n!=0:
    dig=n%10
    rev=rev*10+dig
    n=n//10
print(rev)
```

```python
# WAP TO CHECK WHETHER A GIVEN NUMBER IS AN ARMSTRONG OR NOT?
n=int(input("Enter Any Number:"))
a=len(str(n))
sumdigit=0
arm=n
while n!=0:
    dig =n%10
    sumdigit=sumdigit+ (dig**a)
    n=n//10
if arm==sumdigit:
    print(arm,"ARMSTRONG NUMBER")
else:
    print(arm,"NOT AN ARMSTRONG NUMBER")

```

```python
1-2000
#WAP TO PRINT THE ARMSTORNG NUMBER FROM 1-2000
start=1
end=2000
for i in range(start,end+1):
    a=len(str(i))
    sumdigit=0
    arm=i
    while i!=0:
        dig=i%10
        sumdigit=sumdigit+(dig**a)
        i=i//10
    if arm==sumdigit:
         print(sumdigit,end=" ")    
         
```
