```python
# marks = [85, 72, 90, 45, 67, 38, 95]
# For each student, print grade: A(90+), B(80-89), C(70-79), D(60-69), F(<60)

marks = [85, 72, 90, 45, 67, 38, 95]
for i in range(len(marks)):
    if marks[i] >=90:
        grade="A"
    elif marks[i]>=80:
        grade="B"
    elif marks[i]>=70:
        grade="C"
    elif marks[i]>=60:
        grade="D"
    else:
        grade="F"
    print( "student's grade is ",grade)        
```
