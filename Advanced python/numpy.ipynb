# NUMPY ARRAYS 
##### a numpy array is a collection of elements stored in contiguous memory. It is Designed for numerical computation and scientific computing.
A NumPy array (ndarray) is:
1) A fast
2) Memory-efficient
3) Homogeneous (same data type)

```python
# Write a Python program to get the output --
# l1=[1,6,4,3,5]
# l2=[6,1,3,4,2]
# output=[6,64]

l1=[1,6,4,3,5]
l2=[6,1,3,4,2]
res=[]
fin=[]
for i,j in zip(l1,l2):
    res.append(i**j)
for k in res:
    if k%2==0:
        fin.append(k)
print(fin)
```

```python
l1=[1,6,4,3,5]
l2=[6,1,3,4,2]
res=[]
# output=[6,64]
for i,j in zip(l1,l2):
    if (i**j) %2==0:
        res.append(i**j)
print(res)
```
