```python
import numpy as np
```

# arange function:-

```python
# arange function
arr=np.arange(1,10)
print(arr)
print(arr[0])
print(arr[-1])
print(arr[::])
print(arr[::-1])
print(arr[1::3])
```

```python
arr=np.arange(1,28).reshape(3,3,3)
print(arr)
print("\n")
print(arr[1::2])
print("\n")
print(arr[::-1])
print(arr[0][1][0])
print(arr[2][1][1])
```

# Special Filled Arrays   
By default gives float value
#### 1) np.zeros()
#### 2) np.ones()
#### 3) np.full()
#### 4) np.eye()

### np.zeros(shape)

```python
# np.zeros()
arr=np.zeros(4)
print(arr)
```

```python
arr=np.zeros(4,dtype="int")
print(arr)
```

```python
arr= np.zeros((2,4))
print(arr)
```

```python
arr=np.zeros(((4,4,4)))
print(arr)
```

### np.ones(shape)

```python
arr=np.ones(1)
print(arr)
```

```python
arr=np.ones((2,2))
print(arr)
```

```python
arr=np.ones((2,2,2))
print(arr)
```

### np.full((shape),constant)

```python
# full 
arr=np.full((3,3),5)
print(arr)
```

### np.eye(shape)  
used for diagonal

```python
# eye
arr=np.eye(4,4)
print(arr)
```

```python
arr=np.eye(3,4)
print(arr)
```

```python
arr=np.eye(4,4)
print(arr[::-1])
```

```python
arr=np.array([123,127,126],dtype="int8")
print(arr)
```
