## skewness
#### skewness measures the asymmetry of a probability distribution around its mean, indicating if data points are clustered more towards one side, creating a longer "tail" on the left (negative skew) or right (positive skew). A perfectly symmetrical distribution, like the normal distribution, has zero skewness, where the mean, median, and mode are equal; otherwise, the data is skewed. 

### Types of Skewness
1) <b>Positive Skewness (Right Skew)</b>: The tail extends to the right, meaning most data points are on the lower end, with a few high values pulling the mean up (Mean > Median > Mode).
2) <b>Negative Skewness (Left Skew)</b>: The tail extends to the left, with most data points clustered on the higher end, pulling the mean down (Mean < Median < Mode).
3) <b>Zero Skewness</b>: A perfectly symmetrical distribution, often a normal distribution, where the data is balanced around the center (Mean = Median = Mode). 

## Outliers: abnormal or extreme data points
(Used for null imputation)<br>
Outliers are values in a dataset that are very different from the rest of the data.<br>
They can be unusually high or unusually low compared to most values.

Example:
If most salaries are between 20k–50k and one value is 5 lakhs, that value is an outlier.
### two key rule 
Q1 = 25th percentile                      
Q3 = 75th percentile<br>
 IQR :- Inter Quartile Range
1) IQR :-Q3 − Q1
3) lower fence = q1-1.5*IQR
4) upper fence= q3+1.5*IQR

Values outside this range = outliers

```python
print(np.mean(age))
print(np.median(age))
```

#### using Quantile Function

```python
# By using Quantile 
Q1=np.quantile(age,0.25)
print(Q1)
```

#### using percentile function

```python
# By using Percentile
Q1=np.percentile(age,25)
print(Q1)
```

```python
 Q2=np.quantile(age,0.5)
print(Q2)
```

```python
Q3=np.quantile(age,0.7)
print(Q3)
```

```python
IQR=Q3-Q1
print(IQR)
```

```python
# lower_fence
lower_fence= Q1-(1.5*IQR)
print(lower_fence)
```

```python
# upper_fence
upper_fence=Q3+(1.5*IQR)
print(upper_fence)
```

```python
l=[]
for i in age:
    if i>upper_fence or i<lower_fence:
        l.append(i)
print(l)
```
