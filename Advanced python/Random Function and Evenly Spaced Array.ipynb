## evenly spaced array

#### np.linspace(start,end,[numbers])
#### BY default number value-50


```python

arr=np.linspace(0,10,5)
print(arr)
```

```python
arr=np.linspace(0,10)
print(arr)    # by default 50  numbers 
print(len(arr))
```

## Random function:-

#### randint:
#### random.randint( start , stop,shape)

```python
# Array with random number:
# randint:
# random.randint( start , stop,shape)
arr=np.random.randint(20,30,(2,2,5))
print(arr)
```

```python
# by default take only one value in shape
arr=np.random.randint(20,30)
print(arr)
```

```python
arr=np.random.randint(20,30,5)
print(arr)
```

#### np.random.rand(shape)
#### takes values in between 0 to 1 only 
#### where o is inclusive and 1 is exclusive 

```python
arr=np.random.rand(2,2,5)
print(arr)
```

#### np.random.randn()
#### Takes nearest value to 0 in  both negative and positive integer

```python
arr=np.random.randn(4,6)
print(arr)
```
